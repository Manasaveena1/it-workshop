# Matplotlib

```python
import matplotlib.pyplot as plt
x=np.arange(0,100,0.1)
y=np.sin(x)
fig=plt.figure(figsize=(10,10))
plt.scatter(x,y)
plt.show()
```

# Pie chart

```python
grade={"Ex":20,"A":10,"B":15,"C":12,"D":8}
categories=list(grade.keys())
percentages=list(grade.values())
```

```python
fig=plt.figure(figsize=(6,6))
plt.pie(x=percentages,labels=categories,explode=[0,0,0,0.5,0],shadow=True)
plt.legend(["Ex 90-100%","A 80-90%","B 70-80%","C 60-70%" ,"D 50-60%"])
plt.show()
```

# Bargraph

```python
fig=plt.figure(figsize=(8,8))
plt.bar(x=categories,height=percentages)fig=plt.figure(figsize=(8,8))
plt.xlabel("categories")
plt.ylabel("percentage")
plt.show()
```

```python
fig=plt.figure(figsize=(8,8))
plt.barh(y=categories,width=percentages)
plt.xlabel("categories")
plt.ylabel("percentage")
plt.show()
```
